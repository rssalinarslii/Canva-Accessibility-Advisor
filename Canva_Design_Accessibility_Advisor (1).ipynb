```python
# ============================================
# Canva Design Accessibility Advisor
# ============================================

print("=" * 50)
print("🎨 CANVA DESIGN ACCESSIBILITY ADVISOR")
print("=" * 50)
print("This program helps users create more accessible designs.")
print()
```

```python
def get_design_type():

    while True:

        print("\nSelect Design Type")
        print("1. Poster")
        print("2. Presentation")
        print("3. Instagram Post")
        print("4. Resume")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            return "Poster"

        elif choice == "2":
            return "Presentation"

        elif choice == "3":
            return "Instagram Post"

        elif choice == "4":
            return "Resume"

        else:
            print("Invalid choice. Please try again.")
```

```python
def get_target_audience():

    while True:

        print("\nSelect Target Audience")
        print("1. Children")
        print("2. Students")
        print("3. Adults")
        print("4. Senior Citizens")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            return "Children"

        elif choice == "2":
            return "Students"

        elif choice == "3":
            return "Adults"

        elif choice == "4":
            return "Senior Citizens"

        else:
            print("Invalid choice. Please try again.")
```

```python
def get_background():

    while True:

        print("\nBackground Colour")
        print("1. White")
        print("2. Black")
        print("3. Blue")
        print("4. Yellow")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            return "White"

        elif choice == "2":
            return "Black"

        elif choice == "3":
            return "Blue"

        elif choice == "4":
            return "Yellow"

        else:
            print("Invalid choice. Please try again.")
```

```python
def get_font_size():

    while True:

        try:

            size = int(input("\nEnter Font Size (Example: 18): "))

            if size >= 8 and size <= 72:
                return size

            else:
                print("Please enter a font size between 8 and 72.")

        except ValueError:
            print("Invalid input. Please enter numbers only.")
```

```python
def analyze_accessibility(design_type, audience, background, font_size):

    score = 100
    feedback = []

    # Font Size
    if font_size < 16:
        score -= 20
        feedback.append("Increase font size for better readability.")
    else:
        feedback.append("Font size is appropriate.")

    # Background
    if background == "Yellow":
        score -= 10
        feedback.append("Yellow backgrounds may reduce readability.")

    elif background == "White":
        feedback.append("White background provides good readability.")

    elif background == "Black":
        feedback.append("Black background works well with light text.")

    elif background == "Blue":
        feedback.append("Blue background is suitable with sufficient contrast.")

    # Audience
    if audience == "Senior Citizens":

        if font_size < 20:
            score -= 15
            feedback.append("Senior citizens usually require larger font sizes.")

    elif audience == "Children":
        feedback.append("Use colourful elements and simple language.")

    elif audience == "Students":
        feedback.append("Maintain a clean and organised layout.")

    elif audience == "Adults":
        feedback.append("Use professional colours and balanced spacing.")

    # Design Type
    if design_type == "Presentation":
        feedback.append("Limit text on each slide.")

    elif design_type == "Poster":
        feedback.append("Highlight important information using headings.")

    elif design_type == "Instagram Post":
        feedback.append("Use high contrast for mobile viewing.")

    elif design_type == "Resume":
        feedback.append("Use a clean and professional layout.")

    if score < 0:
        score = 0

    return score, feedback
```

```python
print("\nCollecting User Information...")

design = get_design_type()

audience = get_target_audience()

background = get_background()

font = get_font_size()
```

```python
score, feedback = analyze_accessibility(
    design,
    audience,
    background,
    font
)
```

```python
# Display Accessibility Report

print("\n" + "=" * 50)
print("ACCESSIBILITY REPORT")
print("=" * 50)

print(f"Design Type      : {design}")
print(f"Target Audience  : {audience}")
print(f"Background       : {background}")
print(f"Font Size        : {font}")

print(f"\nAccessibility Score : {score}/100")

if score >= 90:
    rating = "Excellent"

elif score >= 75:
    rating = "Good"

elif score >= 60:
    rating = "Fair"

else:
    rating = "Needs Improvement"

print("Accessibility Rating :", rating)

print("\nRecommendations:")

for item in feedback:
    print("- " + item)
```

```python
while True:

    print("\nWould you like to evaluate another design?")

    answer = input("Enter Y to continue or N to exit: ").upper()

    if answer == "Y":

        design = get_design_type()
        audience = get_target_audience()
        background = get_background()
        font = get_font_size()

        score, feedback = analyze_accessibility(
            design,
            audience,
            background,
            font
        )

        print("\n" + "=" * 50)
        print("ACCESSIBILITY REPORT")
        print("=" * 50)

        print(f"Design Type      : {design}")
        print(f"Target Audience  : {audience}")
        print(f"Background       : {background}")
        print(f"Font Size        : {font}")

        print(f"\nAccessibility Score : {score}/100")

        if score >= 90:
            rating = "Excellent"
        elif score >= 75:
            rating = "Good"
        elif score >= 60:
            rating = "Fair"
        else:
            rating = "Needs Improvement"

        print("Accessibility Rating :", rating)

        print("\nRecommendations:")

        for item in feedback:
            print("- " + item)

    elif answer == "N":

        print("\nThank you for using Canva Design Accessibility Advisor.")
        break

    else:

        print("Invalid choice.")
```
